```python
import numpy as np
import h5py
import cv2
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
import os
```

```python
# for deterministic results
tf.keras.utils.set_random_seed(1)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
os.environ['TF_CUDNN_DETERMINISTIC'] = '1'
```

```python
hdf5_path = 'blooddata-color.hdf5'
```

```python
hdf5_file = h5py.File(hdf5_path, "r")
# Total number of samples
data_num = hdf5_file["train_imgs_resized"].shape[0]

# your train set features
train_set_x_orig = np.array(hdf5_file["train_imgs_resized"][:]) 
# your train set labels
train_set_y_orig = np.array(hdf5_file["train_labels"][:]) 
#training data addresses
train_set_addrs = np.array(hdf5_file["train_addrs"][:])

# your dev set features
dev_set_x_orig = np.array(hdf5_file["dev_imgs_resized"][:]) 
# your train set labels
dev_set_y_orig = np.array(hdf5_file["dev_labels"][:]) 
#training data addresses
dev_set_addrs = np.array(hdf5_file["dev_addrs"][:])

#test data addresses
test_set_addrs = np.array(hdf5_file["test_addrs"][:])
# your test set features
test_set_x_orig = np.array(hdf5_file["test_imgs_resized"][:]) 
# your test set labels
test_set_y_orig = np.array(hdf5_file["test_labels"][:]) 

    
train_set_y_orig = train_set_y_orig.reshape((1, train_set_y_orig.shape[0]))
test_set_y_orig = test_set_y_orig.reshape((1, test_set_y_orig.shape[0]))
dev_set_y_orig = dev_set_y_orig.reshape((1, dev_set_y_orig.shape[0])) #new

train_set_addrs = train_set_addrs.reshape((1, train_set_addrs.shape[0]))
test_set_addrs = test_set_addrs.reshape((1, test_set_addrs.shape[0]))
dev_set_addrs = dev_set_addrs.reshape((1, dev_set_addrs.shape[0])) #New

# Example of a picture
index = 20
# For Grayscale display:
plt.imshow(train_set_x_orig[index])
print(hdf5_file["train_imgs_resized"].shape)
print ("y = " + str(train_set_y_orig[:, index]))
print ("filename = " + str(train_set_addrs[:, index]))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv2D, MaxPool2D, Flatten
from keras.utils import np_utils
```

```python
# building the input vector from the 192x32 pixels
train_set_x_orig = train_set_x_orig.reshape(train_set_x_orig.shape[0], train_set_x_orig.shape[1], train_set_x_orig.shape[2], 3)
test_set_x_orig = test_set_x_orig.reshape(test_set_x_orig.shape[0], test_set_x_orig.shape[1], test_set_x_orig.shape[2], 3)
dev_set_x_orig = dev_set_x_orig.reshape(dev_set_x_orig.shape[0], dev_set_x_orig.shape[1], dev_set_x_orig.shape[2], 3)  #New

train_set_x_orig = train_set_x_orig.astype('float32')
test_set_x_orig = test_set_x_orig.astype('float32')
dev_set_x_orig = dev_set_x_orig.astype('float32')  #New


train_set_x_orig = train_set_x_orig /255
test_set_x_orig = test_set_x_orig /255
dev_set_x_orig = dev_set_x_orig /255  #New
```

```python
# one-hot encoding using keras' numpy-related utilities
n_classes = 7 # i.e., 10 to 90

# 10 -> 1; 20 -> 2; ...; 90 -> 9
train_set_y_orig = train_set_y_orig / 10 - 1
test_set_y_orig = test_set_y_orig /10 - 1
dev_set_y_orig = dev_set_y_orig /10 - 1 #why? New

print("Shape before one-hot encoding: ", train_set_y_orig.shape)
# convert to one-hot
# i.e., 1 -> 100000000; 2 -> 010000000; ...; 9 -> 000000001
train_set_y_orig = np_utils.to_categorical(train_set_y_orig, n_classes)
test_set_y_orig = np_utils.to_categorical(test_set_y_orig, n_classes)
dev_set_y_orig = np_utils.to_categorical(dev_set_y_orig, n_classes)  #New

train_set_y_orig = train_set_y_orig.reshape(train_set_y_orig.shape[1], n_classes)
test_set_y_orig = test_set_y_orig.reshape(test_set_y_orig.shape[1], n_classes)
dev_set_y_orig = dev_set_y_orig.reshape(dev_set_y_orig.shape[1], n_classes)
print("Shape after one-hot encoding: ", train_set_y_orig.shape)
```

```python
#from tensorflow import keras
#model = tf.keras.models.load_model('saved-model-.9627')
```

```python
from tensorflow import keras
print('tensorflow version: ' + tf.__version__)
print('keras version: ' + keras.__version__)
print('numpy version: ' + np.__version__)
print('h5py version: ' + h5py.__version__)
print('cv2 version (opencv-python): ' + cv2.__version__)
print('pandas version: ' + pd.__version__)

# building a linear stack of layers with the sequential model
model = Sequential()
# convolutional layer
model.add(Conv2D(25, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu', input_shape=(100,100,3)))
model.add(Conv2D(35, kernel_size=(3,3), strides=(1,1), padding='valid', activation='relu'))
model.add(MaxPool2D(pool_size=(3,3)))
# flatten output of conv
model.add(Flatten())
# hidden layer
model.add(Dense(100, activation='relu'))
# output layer
model.add(Dense(n_classes, activation='softmax'))

# compiling the sequential model
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')

from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', patience=10)

# training the model for 10 epochs
model.fit(train_set_x_orig, train_set_y_orig, epochs=100, validation_data=(dev_set_x_orig, dev_set_y_orig), callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
losses[['loss', 'val_loss']].plot()
```

```python
from sklearn.metrics import classification_report, confusion_matrix

```

```python
from tensorflow import keras
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
predictions = model.predict(test_set_x_orig)
predictions = to_categorical(np.argmax(predictions,axis=1), num_classes=n_classes)
print(classification_report(test_set_y_orig, predictions))
```

```python
confusion_matrix(np.argmax(test_set_y_orig, axis=1), np.argmax(predictions, axis=1))
```
