```python
from keras_preprocessing.image import ImageDataGenerator
from numpy import expand_dims
from keras_preprocessing.image import load_img, save_img,img_to_array
#from tensorflow.keras.utils import img_to_array

# loop over train addresses
for i in range(len(file_paths)):
   
    # read an image 
    file_path = file_paths[i]
    #print(file_path)
    file_name = file_path.split('\\')[1]
    #print(file_name)
    # cv2 load images as Grayscale
    img = load_img(file_path)
    #print(type(img))
    
    
    # convert image to a numpy array
    img_array = asarray(img) #img_to_array(img)

    samples = expand_dims(img_array, 0)
    #print(samples)
    # create image data augmentation generator
    datagen = ImageDataGenerator(rotation_range=30, fill_mode='constant', cval=255.0)
    
    # prepare iterator
    it = datagen.flow(samples, batch_size=1)
    for j in range(9):
        # generate batch of images
        batch = it.next()
        #print(output_path + file_name + '_rotation_' + str(j)+'.jpg',)
        # save the image with a new filename
        save_img(output_path + '/' +file_name + '_rotation_' + str(j)+'.jpg', batch[0]) #  

#C:/Users/16177/Documents/BloodDropProject/Output/v2/test\10_2021-02-03_impactAngle_10-page-001.jpgParticle_1.tif
```

```python
def createLabels(addrs):
    labels = []
    for addr in addrs:
        if '10_' in addr:
            labels.append(10)
        elif '20_' in addr:
            labels.append(20)
        elif '30_' in addr:
            labels.append(30)
        elif '40_' in addr:
            labels.append(40)
        elif '50_' in addr:
            labels.append(50)
        elif '60_' in addr:
            labels.append(60)
        elif '70_' in addr:
            labels.append(70)
        elif '80_' in addr:
            labels.append(70)
        elif '90_' in addr:
            labels.append(70)
    return labels

#Change addresses to byte object for createLabel function
test_d = './Output/v2/test_rotated/*.jpg'
train_d = './Output/v2/train_rotated/*.jpg'
dev_d = './Output/v2/dev_rotated/*.jpg'

test_dir = glob.glob(test_d)
train_dir = glob.glob(train_d)
dev_dir = glob.glob(dev_d)
hdf5_path = './Output/v2/hdf5/rotated-blooddata-color.hdf5'
```

```python
#Create Labels for images

test_labels = createLabels(test_dir)
train_labels = createLabels(train_dir)
dev_labels = createLabels(dev_dir)
len(train_labels)
```

```python
shuffle_data = True
output_path = 'C:/Users/16177/Documents/BloodDropProject/Output/v2/dev_rotated/*.jpg'
test_output_path = 'C:/Users/16177/Documents/BloodDropProject/Output/v2/test_rotated/*.jpg'
train_output_path = 'C:/Users/16177/Documents/BloodDropProject/Output/v2/train_rotated/*.jpg'

train_addrs = glob.glob(train_output_path)
train_addrs = np.string_(train_addrs)

dev_addrs = glob.glob(output_path)
dev_addrs = np.string_(dev_addrs)

test_addrs = glob.glob(test_output_path)
test_addrs = np.string_(test_addrs)


train_addrs = train_addrs[:]
dev_addrs = dev_addrs[:]
test_addrs = test_addrs[:]
#len(train_addrs)
#len(dev_addrs)
len(test_addrs)
#dev_addrs = addrs[int(0.8*len(addrs)):int(0.9*len(addrs))]

#test_addrs = addrs[int(0.9*len(addrs)):]
```

```python
train_shape = (len(train_addrs), img_height, img_width, 3)
test_shape = (len(test_addrs), img_height, img_width, 3)
dev_shape = (len(dev_addrs), img_height, img_width, 3) # New

# open a hdf5 file and create earrays
hdf5_file = h5py.File(hdf5_path, mode='w')
hdf5_file.create_dataset("train_imgs_resized", train_shape, np.int8)
hdf5_file.create_dataset("test_imgs_resized", test_shape, np.int8)
hdf5_file.create_dataset("dev_imgs_resized", dev_shape, np.int8) #new

#hdf5_file.create_dataset("train_mean", train_shape[1:], np.float32)
hdf5_file.create_dataset("train_labels", (len(train_addrs),), np.int8)
hdf5_file["train_labels"][...] = train_labels
hdf5_file.create_dataset("train_addrs", (len(train_addrs),), dtype="S20")
hdf5_file["train_addrs"][...] = train_addrs
hdf5_file.create_dataset("test_labels", (len(test_addrs),), np.int8)
hdf5_file["test_labels"][...] = test_labels
hdf5_file.create_dataset("test_addrs", (len(test_addrs),), dtype="S20")
hdf5_file["test_addrs"][...] = test_addrs
#New 
hdf5_file.create_dataset("dev_labels", (len(dev_addrs),), np.int8)
hdf5_file["dev_labels"][...] = dev_labels
hdf5_file.create_dataset("dev_addrs", (len(dev_addrs),), dtype="S20")
hdf5_file["dev_addrs"][...] = dev_addrs
```

```python
# a numpy array to save the mean of the images
mean = np.zeros(train_shape[1:], np.float32)
# loop over train addresses
for i in range(len(train_addrs)):
   
    # read an image 
    addr = train_addrs[i]
    #print(str(addr))
    
    # cv2 load images
    img = cv2.imread(addr.decode('utf-8'))
    
    # save the image and calculate the mean so far
    hdf5_file["train_imgs_resized"][i, ...] = img[None]
    mean += img / float(len(train_labels))

# loop over test addresses
for i in range(len(test_addrs)):
   
    # read an image 
    # cv2 load images
    addr = test_addrs[i]
    img = cv2.imread(addr.decode('utf-8'))
    
    # save the image
    hdf5_file["test_imgs_resized"][i, ...] = img[None]

# loop over dev addresses
for i in range(len(dev_addrs)):
   
    # read an image 
    # cv2 load images
    addr = dev_addrs[i]
    img = cv2.imread(addr.decode('utf-8'))
    
    # save the image
    hdf5_file["dev_imgs_resized"][i, ...] = img[None]
# save the mean and close the hdf5 file
#hdf5_file["train_mean"][...] = mean
```
