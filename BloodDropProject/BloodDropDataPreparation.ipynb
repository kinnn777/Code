```python
import random
import glob
import numpy as np
import h5py
import cv2
import os
import pandas as pd
import h5py
```

```python
shuffle_data = True
image_path = 'C:\\Users\\16177\\Documents\\BloodDropProject\\img\\data\\*.jpg'
addrs = glob.glob(image_path)
addrs
```

```python
def createLabels(addrs):
    labels = []
    for addr in addrs:
        if '10_' in addr:
            labels.append(10)
        elif '20_' in addr:
            labels.append(20)
        elif '30_' in addr:
            labels.append(30)
        elif '40_' in addr:
            labels.append(40)
        elif '50_' in addr:
            labels.append(50)
        elif '60_' in addr:
            labels.append(60)
        elif '70_' in addr:
            labels.append(70)
        elif '80_' in addr:
            labels.append(70)
        elif '90_' in addr:
            labels.append(70)
    return labels

labels = createLabels(addrs)
addrs = np.string_(addrs)
#labels
```

```python
if shuffle_data:
    c = list(zip(addrs, labels))
    shuffle(c)
    addrs, labels = zip(*c)
```

```python
test_addrs = addrs[int(0.9*len(addrs)):]
test_labels = labels[int(0.9*len(labels)):]

dev_addrs = addrs[int(0.8*len(addrs)):int(0.9*len(addrs))]
dev_labels = labels[int(0.8*len(labels)):int(0.9*len(labels))]

train_addrs = addrs[0:int(0.8*len(addrs))]
train_labels = labels[0:int(0.8*len(labels))]
```

```python
# address to where you want to save the hdf5 file
hdf5_train_path = r'C:\Users\16177\Documents\BloodDropProject\Output\TrainingData\blooddata_training_data'
hdf5_test_path = r'C:\Users\16177\Documents\BloodDropProject\Output\TestData\blooddata_test_data'
hdf5_dev_path = r'C:\Users\16177\Documents\BloodDropProject\Output\DevData\blooddata_dev_data'
img_width = 100
img_height = 100
```

```python
train_shape = (len(train_addrs), img_height, img_width, 3)
test_shape = (len(test_addrs), img_height, img_width, 3)
dev_shape = (len(dev_addrs), img_height, img_width, 3) # New
```

```python
# open a training hdf5 file and create earrays
hdf5_train_file = h5py.File(hdf5_train_path, mode='w')
hdf5_train_file.create_dataset("train_imgs", train_shape, np.int8)

hdf5_train_file.create_dataset("train_labels", (len(train_addrs),), np.int8)
hdf5_train_file["train_labels"][...] = train_labels
hdf5_train_file.create_dataset("train_addrs", (len(train_addrs),), dtype="S20")
hdf5_train_file["train_addrs"][...] = train_addrs
```

```python
# open a test hdf5 file and create earrays
hdf5_test_file = h5py.File(hdf5_test_path, mode='w')
hdf5_test_file.create_dataset("test_imgs", test_shape, np.int8)

hdf5_test_file.create_dataset("test_labels", (len(test_addrs),), np.int8)
hdf5_test_file["test_labels"][...] = test_labels
hdf5_test_file.create_dataset("test_addrs", (len(test_addrs),), dtype="S20")
hdf5_test_file["test_addrs"][...] = test_addrs
```

```python
# open a test hdf5 file and create earrays
hdf5_dev_file = h5py.File(hdf5_dev_path, mode='w')
hdf5_dev_file.create_dataset("dev_imgs", dev_shape, np.int8)

hdf5_dev_file.create_dataset("dev_labels", (len(dev_addrs),), np.int8)
hdf5_dev_file["dev_labels"][...] = dev_labels
hdf5_dev_file.create_dataset("dev_addrs", (len(dev_addrs),), dtype="S20")
hdf5_dev_file["dev_addrs"][...] = dev_addrs
```

```python

for i in range(len(train_addrs)):
    #filename = str(train_addrs[i]).replace("b'","")#read image
    ##print(filename)
    im = train_addrs[i]
    img = cv2.imread(im.decode('utf-8'))#load image
    #cv2.imwrite(filename,img)
    hdf5_train_file["train_imgs"][i, ...] = img[None]

hdf5_train_file.close()
```

```python
for i in range(len(test_addrs)):
    #filename = str(train_addrs[i]).replace("b'","")#read image
    ##print(filename)
    im = test_addrs[i]
    img = cv2.imread(im.decode('utf-8'))#load image
    hdf5_test_file["test_imgs"][i, ...] = img[None]

hdf5_test_file.close()
```

```python
for i in range(len(dev_addrs)):
    #filename = str(train_addrs[i]).replace("b'","")#read image
    ##print(filename)
    im = dev_addrs[i]
    img = cv2.imread(im.decode('utf-8'))#load image
    hdf5_dev_file["dev_imgs"][i, ...] = img[None]

hdf5_dev_file.close()
```
